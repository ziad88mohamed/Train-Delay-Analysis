#### Data Cleaning & Preprocessing

```python
rail.columns
```

```python
#  (Date of Purchase, Date of Journey)  converted to datetime format.

rail[['Date of Purchase', 'Date of Journey']] =rail[['Date of Purchase', 'Date of Journey']].apply(pd.to_datetime, errors='coerce')
```

```python
# (Time of Purchase, Departure Time, Arrival Time, Actual Arrival Time)  converted to time format.

rail['Actual Arrival Time'] = rail['Actual Arrival Time'].fillna("00:00:00")
rail['Actual Arrival Time'] = pd.to_datetime(rail['Actual Arrival Time']).dt.time
rail['Actual Arrival Time'] = pd.to_datetime(rail['Actual Arrival Time'].astype(str), format='%H:%M:%S').dt.time
rail['Arrival Time'] = pd.to_datetime(rail['Arrival Time'].astype(str), format='%H:%M:%S').dt.time
rail['delay'] = (pd.to_datetime(rail['Actual Arrival Time'].astype(str), format='%H:%M:%S') -
             pd.to_datetime(rail['Arrival Time'].astype(str), format='%H:%M:%S')).dt.total_seconds()
rail['delay'] = pd.to_timedelta(rail['delay'], unit='s')
```

```python
rail.isnull().sum()
```

# Missing Values



```python
rail['Railcard'].unique()
```

```python
rail["Railcard"].fillna('None', inplace=True)
rail["Reason for Delay"].fillna('No Dellay', inplace=True)
```

```python
rail.isnull().sum()
```

```python
rail.duplicated().sum()
```

```python
rail.drop(['Transaction ID'],axis=1,inplace=True)
```

```python
rail["Booking Lead Time"] = (rail["Date of Journey"] - rail["Date of Purchase"]).dt.days
```

```python
rail['Departure Time'] = pd.to_datetime(rail['Departure Time'], format='%H:%M:%S', errors='coerce')
rail['Actual Arrival Time'] = pd.to_datetime(rail['Actual Arrival Time'], format='%H:%M:%S', errors='coerce')

# Compute duration (arrival - departure)
rail['Journey Duration'] = rail['Actual Arrival Time'] - rail['Departure Time']

# Convert timedelta to hours & minutes
rail['Journey Duration'] = rail['Journey Duration'].astype(str)  # Optional: Convert to string format
```

```python
rail['Journey Duration'] = rail['Journey Duration'].astype(str).str.split().str[-1]
```

```python
rail['delay'] = rail['delay'].astype(str).str.split().str[-1]
```

```python
rail['Actual Arrival Time'] = rail['Actual Arrival Time'].astype(str).str.split().str[-1]
```

```python
rail['Departure Time'] = rail['Departure Time'].astype(str).str.split().str[-1]
```

```python
rail.head()
```

```python

# Convert Date columns to datetime format
rail['Date of Purchase'] = pd.to_datetime(rail['Date of Purchase'], errors='coerce')
rail['Date of Journey'] = pd.to_datetime(rail['Date of Journey'], errors='coerce')

# Convert Time columns to proper time format (12-hour AM/PM)
time_columns = ['Time of Purchase', 'Departure Time', 'Arrival Time', 'Actual Arrival Time']
for col in time_columns:
    rail[col] = pd.to_datetime(rail[col], format='%H:%M:%S', errors='coerce').dt.strftime('%I:%M:%S %p')

# Convert Price and Booking Lead Time to integer (if not already)
rail['Price'] = pd.to_numeric(rail['Price'], errors='coerce').fillna(0).astype(int)
rail['Booking Lead Time'] = pd.to_numeric(rail['Booking Lead Time'], errors='coerce').fillna(0).astype(int)

# Convert Journey Duration to timedelta and extract only time (HH:MM:SS)
rail['Journey Duration'] = pd.to_timedelta(rail['Journey Duration'], errors='coerce')
rail['Journey Duration'] = rail['Journey Duration'].astype(str).str.split().str[-1]  # Keep only HH:MM:SS

# Check updated data types
print(rail.dtypes)
```

```python


# List of time columns
time_columns = ['Time of Purchase', 'Departure Time', 'Arrival Time', 'Actual Arrival Time']

# Convert time columns to proper time format (HH:MM:SS)
for col in time_columns:
    rail[col] = pd.to_datetime(rail[col], format='%H:%M:%S', errors='coerce').dt.time

# Verify data types
print(rail.dtypes)
```

#### EDA


```python
# Set style
sns.set_style("whitegrid")

# 1. Ticket Price Distribution
plt.figure(figsize=(8, 5))
sns.histplot(rail['Price'], bins=30, kde=True)
plt.title("💰 Ticket Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# Ensure 'Journey Duration' is properly converted to timedelta
rail['Journey Duration'] = pd.to_timedelta(rail['Journey Duration'], errors='coerce')

# Now, create the histogram correctly
plt.figure(figsize=(8, 5))
sns.histplot(rail['Journey Duration'].dropna().dt.total_seconds() / 60, bins=30, kde=True)
plt.title("⏳ Journey Duration Distribution (Minutes)")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Count")
plt.show()
```

```python
# --- 4. Delays & Reasons for Delay ---
plt.figure(figsize=(10, 5))
sns.countplot(y=rail['Reason for Delay'], order=rail['Reason for Delay'].value_counts().index, palette="coolwarm")
plt.title("⚠️ Reasons for Delays")
plt.xlabel("Count")
plt.ylabel("Reason for Delay")
plt.show()
```

```python
# --- 5. Refund Requests ---
plt.figure(figsize=(6, 4))
sns.countplot(x=rail['Refund Request'], palette="Set2")
plt.title("💰 Refund Request Distribution")
plt.xlabel("Refund Requested")
plt.ylabel("Count")
plt.show()
```

```python
# --- 6. Popular Departure & Arrival Stations ---
plt.figure(figsize=(10, 5))
sns.barplot(y=rail['Departure Station'].value_counts().index[:10],
            x=rail['Departure Station'].value_counts().values[:10],
            palette="viridis")
plt.title("📍 Top 10 Departure Stations")
plt.xlabel("Count")
plt.ylabel("Departure Station")
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(y=rail['Arrival Destination'].value_counts().index[:10],
            x=rail['Arrival Destination'].value_counts().values[:10],
            palette="magma")
plt.title("📍 Top 10 Arrival Destinations")
plt.xlabel("Count")
plt.ylabel("Arrival Destination")
plt.show()
```

```python
# --- 1. Box Plot of Ticket Prices (Detect Outliers) ---
plt.figure(figsize=(8, 5))
sns.boxplot(x=rail['Price'])
plt.title("💰 Box Plot of Ticket Prices (Outlier Detection)")
plt.xlabel("Price")
plt.show()
```

```python

# --- 2. Average Journey Duration by Ticket Class ---
plt.figure(figsize=(8, 5))
sns.barplot(x=rail['Ticket Class'],
            y=rail['Journey Duration'].dt.total_seconds() / 60,
            estimator=lambda x: sum(x) / len(x),
            palette="Blues")
plt.title("⏳ Average Journey Duration by Ticket Class")
plt.xlabel("Ticket Class")
plt.ylabel("Avg Duration (Minutes)")
plt.show()

```

```python
# --- 4. Most Common Routes (Departure → Arrival) ---
rail['Route'] = rail['Departure Station'] + " → " + rail['Arrival Destination']
top_routes = rail['Route'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_routes.index, x=top_routes.values, palette="coolwarm")
plt.title("🚆 Top 10 Most Common Train Routes")
plt.xlabel("Number of Trips")
plt.ylabel("Route")
plt.show()
```

```python

# --- 5. Ticket Sales Trend Over Time ---
plt.figure(figsize=(10, 5))
rail.groupby(rail['Date of Purchase'].dt.date).size().plot()
plt.title("📅 Ticket Purchases Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=45)
plt.show()

```

```python
# --- 6. Payment Methods Distribution ---
plt.figure(figsize=(6, 4))
sns.countplot(x=rail['Payment Method'], palette="Set1")
plt.title("💳 Distribution of Payment Methods")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```
